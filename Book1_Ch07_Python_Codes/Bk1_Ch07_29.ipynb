## Q1. 给定一个整数列表 [3, 5, 2, 7, 1]，找到其中的最大值和最小值，并打印两者之和。

```python
num_list = [3, 5, 2, 7, 1]
max_num = num_list[0]
min_num = num_list[0]

for num in num_list:
    if num > max_num:
        max_num = num
    elif num < min_num:
        min_num = num

print(max_num + min_num)
```

## Q2. 使用 while 循环输出 1 到 10 的所有奇数。

```python
num = 1
while num <= 10:
    if num % 2 != 0:
        print(num)
    num += 1
```

## Q3. 输入一个数字并将其转换为整数，如果输入的不是数字，则提示用户重新输入直到输入数字为止。

```python
while True:
    try:
        num = int(input("请输入一个数字："))
        break
    except ValueError:
        print("输入有误，请重新输入！")

print("输入的整数是：", num)
```

## Q4. 求 100 以内的素数。

```python
prime_list = []

for num in range(2, 101):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        prime_list.append(num)

print(prime_list)
```

## Q5. 请用至少两种不同办法计算1-100中奇数之和。

```python
sum_odd_numbers = 0
for i in range(1, 101):
    if i % 2 != 0:
        sum_odd_numbers += i
print(sum_odd_numbers)
```

```python
sum_odd_numbers = 0
i = 1
while i <= 100:
    if i % 2 != 0:
        sum_odd_numbers += i
    i += 1
print(sum_odd_numbers)
```

```python
sum_odd_numbers = sum([i for i in range(1, 101) if i % 2 != 0])
print(sum_odd_numbers)
```

```python
sum_odd_numbers = sum([i for i in range(1, 101, 2)])
print(sum_odd_numbers)
```

```python
sum_odd_numbers = sum(list(range(1, 101, 2)))
sum_odd_numbers
```

```python
sum_odd_numbers = 0
i = 1
while i <= 100:
    sum_odd_numbers += i
    i += 2
print(sum_odd_numbers)
```

```python
numbers = list(range(1, 101))
sum_odd_numbers = sum(numbers[::2])
print(sum_odd_numbers)
```

```python
import numpy as np

numbers = np.arange(1, 101)
sum_odd_numbers = np.sum(numbers[numbers % 2 != 0])
print(sum_odd_numbers)
```
